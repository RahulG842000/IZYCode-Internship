# correlation matrix

```python
selected_columns = ['Total Doses Administered', 'Male(Individuals Vaccinated)', 'Female(Individuals Vaccinated)']
```

```python
columns_for_analysis = ['Male(Individuals Vaccinated)', 'Female(Individuals Vaccinated)' , "18-44 Years(Individuals Vaccinated)" , "45-60 Years(Individuals Vaccinated)"]

# Perform correlation analysis on selected columns
selected_columns_vaccine = vaccine[columns_for_analysis]
correlation_matrix = selected_columns_vaccine.corr()

# Visualize correlations using a heatmap
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
# we see male and female indivual vacinated is 1.0 same scale
# 45-60 years indvisual vacinated scale is 1.00 is highest age vacinated
```

```python
plt.figure(figsize=(10, 6))
plt.plot(covid19.index, covid19['Confirmed'], marker='o', linestyle='-')
plt.title('Daily COVID-19 Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Confirmed')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
